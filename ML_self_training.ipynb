```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python
k = 5
n = 1000
chi = sts.chi2(k) # задаем распределение с параметром k
sample = chi.rvs(n) # генерируем выборку объемом n
#print(sample)
x = np.linspace(0, 30, 200)
plt.hist(sample, density=True, bins = 20, histtype='bar', ec='black')
plt.plot(x, chi.pdf(x))
plt.title('Гистограмма выборки и теоретическая плотность распределения')
plt.xlabel('$x$')
```

```python
k = 5
n = 5
chi = sts.chi2(k) # задаем распределение с параметром k
sample = chi.rvs(n) # генерируем выборку объемом n
#print(sample)
sample_mean_arr = [] # создаем список для хранения выборочных средних
for i in range(1000): # цикл по выборкам
    sample = chi.rvs(n) # генерируем выборку объема n
    sample_mean = sum(sample) / n # вычисляем выборочное среднее как среднее арифметическое
    sample_mean_arr.append(sample_mean) # добавляем полученное значение в список

sample_mean_arr = np.array(sample_mean_arr) # переходим от списка к массиву numpy
plt.hist(sample_mean_arr, density=True, bins = 20, histtype='bar', ec='black') # строим гистограмму по выборке выборочных средних
plt.title('Гистограмма распределения выборочных средних при $n=5$ и \n плотность нормального распределения с соответствующими параметрами (по ЦПТ)')

norm = sts.norm(k, np.sqrt(2*k / n)) # создаем случайную нормально распределенную величину
x = np.linspace(0, 20, 100)
plt.plot(x, norm.pdf(x))
plt.xlabel('x')
```
